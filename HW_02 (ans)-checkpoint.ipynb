```python
import pandas as pd
import numpy as np

from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.manifold import TSNE

import warnings
warnings.simplefilter('ignore')
```

## 讀取檔案

```python
data = pd.read_csv('./train.csv')
display(data.head())
```

```python
display(data.info())
```

## 取出連續型特徵 (方便降維)

```python
data_conti = data.select_dtypes(include=['int64','float64']).drop(columns='Id')
data_conti.head()
```

## 分析缺失值

```python
data_null = data_conti.iloc[:,np.where(data_conti.isnull().sum() != 0)[0]]
display(data_null.head())

index = np.arange(len(data_null.columns))
bar = plt.bar(index, data_null.isnull().sum(), width=0.5, alpha=0.8, color='coral')
for item in bar:
    height = item.get_height()
    plt.text(item.get_x() + 0.25, height * 1.05, '%d' % int(height),
             ha='center', va='bottom', fontsize=13)
plt.ylabel('Null', fontsize=13)
plt.xlabel('Feature', fontsize=13)
plt.title('Number of Nulls', fontsize=13)
plt.xticks(index, data_null.columns)
plt.grid(True)
plt.show()
```

## 去除缺失值

```python
data_conti = data_conti.dropna()
```

## 轉化成 ndarray 後進行標準化，並將 Y 透過四分位數轉換成四個類別

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_z = scaler.fit_transform(data_conti)
data_z = pd.DataFrame(data_z, columns=data_conti.columns)
display(data_z.head())
```

```python
data_z['label'] = pd.qcut(data_z.SalePrice, 4, labels=[1, 2, 3, 4])
data_z.head()
```

## 利用 TSNE 降維

```python
tsne = TSNE(
    n_components=2, init='pca', perplexity = 40, 
    early_exaggeration = 8, learning_rate = 150,
    n_iter = 2000
)
X_tsne = tsne.fit_transform(data_z.iloc[:,:36])
```

```python
df_tsne = pd.DataFrame(X_tsne, columns=['X_tsne1', 'X_tsne2'])
df_tsne['label'] = data_z.label
display(df_tsne.head())
```

## 做圖

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(data=df_tsne, x='X_tsne1', y='X_tsne2', hue='label', ax=ax, alpha=0.4)
ax.legend(loc='upper left')
plt.show()
```

## 請問同學從降維後的圖表中可以看出什麼現象？
label對應的SalePrice : 4 > 3 > 2 > 1 
1. label3 & label4 、 label1 & label2 分別明顯集中在右、左邊，1&2、3&4兩組之間分的蠻明確(雖然各有一些混雜，但先不討論離群值)。
2. 1&2、3&4兩組雖然分得很清楚，但組內的重合性卻高

## 請同學分析，什麼因素導致降維後的點座標重合性高？
因為最後分出的label是將SalePrice切割後得出的分類，在切割點附近會有許多較模糊的資料容易重疊，造成最後在視覺化上的重合性。能看到目標變數SalePrice是明顯的right-skewd，導致左邊集中區塊的類別較不易區分。

```python
plt.figure(figsize=(25, 6))

plt.subplot(131)
sns.distplot(data_conti.SalePrice, kde=True)
plt.title('Kernel Density Estimation')

plt.subplot(132)
sns.distplot(data_conti.SalePrice, fit=stats.johnsonsu, kde=False)
plt.title('John SU distribution')

plt.subplot(133)
sns.distplot(data_conti.SalePrice, kde=False, fit=stats.norm)
plt.title('Normal Distribution')

plt.show()
```

```python
def plot_sp(x, y, x_label, y_label, fontsize=12):
    plt.scatter(x, y)
    plt.xlabel(x_label, fontsize=fontsize)
    plt.ylabel(y_label, fontsize=fontsize)
y = data_conti.SalePrice
y_label='SalePrice'
```

```python
plt.figure(figsize=(30, 45), dpi=100)
for i, f in enumerate(data_conti.columns[:-1]):
    x_label = f
    plt.subplot(9, 4, i+1)
    plot_sp(data_conti[x_label],
            y,
            x_label,
            y_label,
            fontsize=10)
plt.show()
```

```python
corr, p_val = stats.spearmanr(data_conti, nan_policy='omit')

plt.figure(figsize=(20, 8), dpi=100)

plt.subplot(121)
plt.title('Spearman Correlation Heatmap')
sns.heatmap(corr, xticklabels=data_conti.columns, yticklabels=data_conti.columns)

plt.subplot(122)
plt.title('P Value Heatmap')
sns.heatmap(p_val, xticklabels=data_conti.columns, yticklabels=data_conti.columns)

plt.show()
```

能看到其實連續型的變數很多，但並不是每個變數與SalePrice的相關係數都是高的，把所有連續型變數都加進去的話難免會影響到降維後的視覺化，因為摻雜的訊息太多了，下面我試著把相關係數超過0.6的變數取出。

```python
corr6 = []
for i in range(len(data_z.columns)-2): #不要 label 和 SalePrice
    if np.abs(data_z.corr().iloc[i,-1])>0.6:
        corr6.append(data_z.columns[i])
data6 = data_z[corr6]
data6['label'] = data_z.label
```

拿上面的變數再做一次降維並視覺化。

```python
tsne = TSNE(
    n_components=2, init='pca', perplexity = 40, 
    early_exaggeration = 8, learning_rate = 150,
    n_iter = 2000
)
X_tsne6 = tsne.fit_transform(data_z[corr6])
X_tsne6 = pd.DataFrame({'tsne1' : X_tsne6[:,0],
                        'tsne2' : X_tsne6[:,1],
                        'label' : data_z.label})
X_tsne6.head()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
sns.scatterplot(data=X_tsne6, x='tsne1', y='tsne2', hue='label', ax=ax, alpha=0.4)
ax.legend(loc='best')
plt.show()
```
