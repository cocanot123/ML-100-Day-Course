## 讀取資料

```python
""" Your code hear: 讀取 train.csv, test.csv 並合併 """
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
df = pd.concat((train, test), axis=0)
```

## 繪圖

```python
# 繪圖
""" Your code hear: """
missing = df.isnull().sum(axis = 0)
missing = missing[missing > 0]
missing.sort_values(inplace=True, ascending = False)

plt.figure(figsize = ( 10,5))
plt.bar(list(missing.index), np.array(missing))
plt.xticks(rotation='vertical')
plt.show()
```

```python
def plotScat(x1, y1, x2, y2):
    plt.figure(figsize = (16, 6), dpi = 70)
    plt.subplot(1, 2, 1, title = 'Train')
    plt.scatter(x1, y1)
    plt.subplot(1, 2, 2, title = 'Test')
    plt.scatter(x2, y2)

def plotHist(x1,x2):
    plt.figure(figsize = (16, 6), dpi = 70)
    plt.subplot(1,2,1, title = 'Train')
    sns.histplot(x1, kde = True, stat = 'probability', element = 'bars')
    plt.subplot(1,2,2, title = 'Test')
    sns.histplot(x2, kde = True, stat = 'probability', element = 'bars')
    

def nan_analysis(feature, target = 'SalePrice', na = 'Na'):
    x1 = train[feature].fillna(na)
    y1 = train[target]
    x2 = test[feature].fillna(na)
    y2 = test[target]
    plotScat(x1, y1, x2, y2)
    plotHist(x1,x2)
```

```python
#Average
feature = 'LotFrontage'
nan_analysis(feature, na = train[feature].mean())
feature = 'GarageYrBlt'
nan_analysis(feature, na = train[feature].mean())
feature = 'MasVnrArea'
nan_analysis(feature, na = train[feature].mean())
```

```python
#Median
feature = 'LotFrontage'
nan_analysis(feature, na = train[feature].median())
feature = 'GarageYrBlt'
nan_analysis(feature, na = train[feature].median())
feature = 'MasVnrArea'
nan_analysis(feature, na = train[feature].median())
```

## 分析標準差

```python
""" Your code hear: """
feature = 'LotFrontage'
std_original = pd.DataFrame.std(train[feature])
std_mean = pd.DataFrame.std(train[feature].fillna(train[feature].mean()))
std_median = pd.DataFrame.std(train[feature].fillna(train[feature].median()))
print('LotFrontage mean: ' + str(std_mean), 'LotFrontage original: ' + str(std_original), 'LotFrontage median: ' + str(std_median))

feature = 'GarageYrBlt'
std_original = pd.DataFrame.std(train[feature])
std_mean = pd.DataFrame.std(train[feature].fillna(train[feature].mean()))
std_median = pd.DataFrame.std(train[feature].fillna(train[feature].median()))
print('GarageYrBlt mean: ' + str(std_mean), 'GarageYrBlt original: ' + str(std_original), 'GarageYrBlt median: ' + str(std_median))

feature = 'MasVnrArea'
std_original = pd.DataFrame.std(train[feature])
std_mean = pd.DataFrame.std(train[feature].fillna(train[feature].mean()))
std_median = pd.DataFrame.std(train[feature].fillna(train[feature].median()))
print('MasVnrArea mean: ' + str(std_mean), 'MasVnrArea original: ' + str(std_original), 'MasVnrArea median: ' + str(std_median))
```
